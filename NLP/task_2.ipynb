```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.datasets import load_files
from gensim import corpora, models
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
bbc_data = load_files('./../bbc')
X, y = bbc_data.data, bbc_data.target
```

```python
documents = []

stemmer = WordNetLemmatizer()

for sen in range(0, len(X)):

    document = re.sub(r'\W', ' ', str(X[sen]))
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document)
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    document = re.sub(r'^b\s+', '', document)
    document = document.lower()
    document = document.split()
    document = [stemmer.lemmatize(word) for word in document]
    document = ' '.join(document)

    documents.append(document)
```

```python
lemmas = documents
```

```python
stop_words=stopwords.words('english')
texts = [[word for word in document.lower().split() if word not in stop_words] for document in documents]
```

```python
business = []
entertainment = []
politics = []
sport = []
tech = []

for i, categ in enumerate(y):
    if categ == 0:
        business.append(texts[i])
    elif categ == 1:
        entertainment.append(texts[i])
    elif categ == 2:
        politics.append(texts[i])
    elif categ == 3:
        sport.append(texts[i])
    elif categ == 4:
        tech.append(texts[i])
```

```python
label = business
dictionary = corpora.Dictionary(label)
corpus = [dictionary.doc2bow(text) for text in label]
lda = models.LdaModel(corpus=corpus,
                      id2word=dictionary,
                      num_topics=15,
                      alpha='auto',
                      eta='auto',
                      iterations = 20,
                      passes = 5)
vis_data = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```
